```python
def get_adj(i,n):
    adj = []
    if i <= N: #the top
        if i % N == 1: #on the left
            if i+1 <= length  : adj.append(loc(i+1,n, 1))
            if i+N <= length  : adj.append(loc(i+N,n, 1))
            if i+N+1 <= length: adj.append(loc(i+N+1,n, 1))
        elif i % N == 0: #on the right
            if 1 <= i-1 : adj.append(loc(i-1,n, -1))
            if i+N-1 <= length  : adj.append(loc(i+N-1,n, 1))
            if i+N <= length  : adj.append(loc(i+N,n, 1))
        else: #in the middle
            if 1 <= i-1 : adj.append(loc(i-1,n, -1))
            if i+N-1 <= length  : adj.append(loc(i+N-1,n, 1))
            if i+N <= length  : adj.append(loc(i+N,n, 1))
            if i+N+1 <= length: adj.append(loc(i+N+1,n, 1))
            if i+1 <= length  : adj.append(loc(i+1,n, 1))
            
            
        
        
    elif  length - N < i: #the bottom
        if i % N == 1: #on the left
            if i+1 <= length  : adj.append(loc(i+1,n, 1))
            if i - N >= 1 : adj.append(loc(i - N, n, -1))
            if i - N +1>= 1 : adj.append(loc(i - N+1, n, -1))
        elif i % N == 0: #on the right
            if i-1 >= 1  : adj.append(loc(i-1,n, -1))
            if i - N >= 1 : adj.append(loc(i - N, n, -1))
            if i - N-1>= 1 : adj.append(loc(i - N-1, n, -1))
        else: #in the middle
            if i-1 >= 1  : adj.append(loc(i-1,n, -1))
            if i - N-1>= 1 : adj.append(loc(i - N-1, n, -1))
            if i+1 <= length  : adj.append(loc(i+1,n, 1))
            if i - N +1>= 1 : adj.append(loc(i - N+1, n, -1))
            
            if i - N >= 1 : adj.append(loc(i - N, n, -1))
            

    else:  # N < i <= length - N #the middle
        if i % N == 1: #on the left
            if i - N >= 1 : adj.append(loc(i - N, n, -1))
            if i - N +1>= 1 : adj.append(loc(i - N+1, n, -1))
            if i+1 <= length  : adj.append(loc(i+1,n, 1))
            if i+N <= length  : adj.append(loc(i+N,n, 1))
            if i+N+1 <= length: adj.append(loc(i+N+1,n, 1))
            
        elif i % N == 0: #on the right
            if 1 <= i-1 : adj.append(loc(i-1,n, -1))
            if i+N-1 <= length  : adj.append(loc(i+N-1,n, 1))
            if i+N <= length  : adj.append(loc(i+N,n, 1))
            if i - N >= 1 : adj.append(loc(i - N, n, -1))
            if i - N-1>= 1 : adj.append(loc(i - N-1, n, -1))
        else: #in the middle
            if i - N-1>= 1 : adj.append(loc(i - N-1, n, -1))
            if i - N >= 1 : adj.append(loc(i - N, n, -1))
            if i - N +1>= 1 : adj.append(loc(i - N+1, n, -1))
            if i-1 >= 1  : adj.append(loc(i-1,n, -1))   
            if i+1 <= length  : adj.append(loc(i+1,n, 1))
            if i+N-1 <= length  : adj.append(loc(i+N-1,n, 1))
            if i+N <= length  : adj.append(loc(i+N,n, 1))
            if i+N+1 <= length: adj.append(loc(i+N+1,n, 1))
            
            
    
    return adj
```
