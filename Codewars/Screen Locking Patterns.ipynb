```python
class graph:
    def __init__(self):
        self.adj = [
                [2,4,5,6,8],
                [1,3,4,5,6,7,9],
                [2,4,5,6,8],
                [1,2,3,5,7,8,9],
                [1,2,3,4,6,7,8,9],
                [1,2,3,5,7,8,9],
                [2,4,5,6,8],
                [1,3,4,5,6,7,9],
                [2,4,5,6,8],
            ]
        self.exceptions = {
            (1,2):3,
            (4,5):6,
            (7,8):9,
            (1,4):7,
            (2,5):8,
            (3,6):9,
            (1,5):9,
            (3,5):7,
            (3,2):1,
            (6,5):4,
            (9,8):7,
            (7,4):1,
            (8,5):2,
            (9,6):3,
            (9,5):1,
            (7,5):3,
        }
        
        
    def test_cross(self, a, b):
        if (a,b) in self.exceptions:
                return self.exceptions[(a,b)]
        return False
    
    def adjacent(self, a):
        return self.adj[a-1]
        
```

```python
Patterns = graph()
```

```python
def parse(start, length, crossed):
    if length == 1 :
        #print(crossed + [start])
        return 1
    else:
        L = Patterns.adjacent(start)
        #print(start, L)
        crossed.append(start)
        c = 0
        for p in L:
            if not p in crossed:
                #print(p)
                c += parse(p, length-1, crossed.copy())
            else:
                #print("ON")
                v = Patterns.test_cross(start,p)
                if v != False:
                    if not v in crossed:
                        c += parse(v, length-1, crossed.copy())
                """else:
                    print("OFF")"""
        return c
```
