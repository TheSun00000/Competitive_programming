```python
def printGrid(grid):
    res = ""
    for row in grid:
        for col in row[:-1]:
            print(col, end='')

class Node:
    def __init__(self, x, y) -> None:
        self.children = []
        self.x = x
        self.y = y
    def addChild(self, node):
        self.children.append(node)
        

def explore(node):
    x, y = node.x, node.y
    printGrid(grid)
    print('------------')

    if grid[x][y] == 't':
        return True
    else:
        grid[x][y] = 'v'
        
    for dx, dy in zip(dxs, dys):
            if 0<=x+dx<H and 0<=x+dy<W and not grid[x+dx][y+dy] in ['x','v'] :    
                next_node = Node(x+dx, y+dy)
                node.addChild(next_node)
                r = explore(next_node)
                if r:return True

    return False

def chercherLaPersonne(grid):
    for i in range(len(grid)):
        for j in range(len(grid[0])):
            if grid[i][j] == 'p':
                return i, j



# Un exemple pour le tester:
grid = """xxxxxxxxxxxxxxxxxxxx
x      x      x  p x
x      x           x
x      xxxxxxxxx   x
x              x   x
xxxxxxxxxx     x   x
xt                 x
xxxxxxxxxxxxxxxxxxxx"""

dxs = [0,0,1,-1]
dys = [1,-1,0,0]
H, W = len(grid), len(grid)

grid = grid.split('\n')
grid = [list(line)+['\n', '\0'] for line in grid]

x, y = chercherLaPersonne(grid)



#x et y sont les coordonnes de P
node = Node(x, y)
trouve = explore(node)
if trouve:
    print('TRESOR TROUVE')
else:
    print('TRESOR NON TROUVE')
```
